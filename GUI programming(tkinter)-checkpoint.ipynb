**1. creating Label:**
 
 Lebel is a predefine class.
 Labels are used to display either text or image.
Syntax:
      l=Label(master,option,..)

```python
from tkinter import *
window=Tk()
label=Label(window,text="Say Hello")
label.pack()
window.mainloop()

```

**2. Creating Layout/geometry:**

 The process of arrangement of components(Widgets on the window to implement GUI programming)
 We can arrange the components using three predefine layouts)
 1. Pack layout
 2. Grid layout
 3. place layout
 

```python
# Example:
from tkinter import *
window=Tk()
logo=PhotoImage(file='hello.gif')
label=Label(window,image=logo)
label.pack()
```

3. Creating Canvas:

The canvas is a rectangular area intended for drawing pictures or other complex layouts.
you can place graphics,text,widgets or frames on a canvas.

Syntax:
       w=Canvas(master,option=value,...)

```python
from tkinter import *
h = 190
w =150
master = Tk()
w = Canvas(master, width=w, height=h)
w.pack()
w.create_oval(50,50,100,100)
w.create_rectangle(100,100,50,50)
master.mainloop()
```

**4. Creating frame:**

The frame widget is very important for the process of grouping and organizing other widgets in a somehow.
It uses rectangular areas in the screen to organize the layout and to provide padding of these widgets.
A frame can also be used as a foundation class to implement complex widgets.

Syntax:
      w=Frame(master,option,...)

```python
from tkinter import *
root=Tk()

frame=Frame(root)
frame.pack()

bottomframe=Frame(root)
bottomframe.pack(side=BOTTOM)

redbutton=Button(frame,text="Red",fg="red")
redbutton.pack(side=LEFT)

greenbutton=Button(frame,text="Brown",fg="brown")
greenbutton.pack(side=LEFT)

bluebutton=Button(frame,text="Blue",fg="blue")
bluebutton.pack(side=LEFT)

blackbutton=Button(frame,text="Black",fg="black")
blackbutton.pack(side=BOTTOM)

root.mainloop()
```

**5. Creating Spinbox:**

The spinbox widgets is a variant of the standard Tkinter Entry widgets,
which can be used to select from a fixed number of values.

Syntax:
       w=Spinbox(master,option,...)


```python
# Example:
from tkinter import *
m=Tk()
w=Spinbox(master,from=0,to 10)
w.pack()
m.mainloop()
```

6. Creating checkbutton:

The checkbutton widget is used to display a number of options to user as toggle buttons.
The user can then select one or more options by clicking the button corresponding to each option.

Syntax:
       w=Checkbutton(master,option,...)
       

```python
# Example:
from tkinter import *
top=Tk()
v1=IntVar()
v2=IntVar()
c1=Checkbutton(top,text="Music",variable=v1)
c2=Checkbutton(top,text="Video",variable=v2)
c1.pack()
c2.pack()
top.mainloop()
```

7. Creating Listbox:
The listbox widgets is used to display a list of items from which a user can select a number of items.

Syntax: 
       w=Listbox(master,option,...)

```python
# Example:
from tkinter import *

top=Tk()
l=Listbox(top)
l.insert(1,"Java")
l.insert(2,"Javascript")
l.insert(3,"php")
l.insert(4,"python")
l.insert(5,"c")
l.pack()
top.mainloop()
```

8. Creating Radiobutton:

This widgets implement a multiple-choice buttton, which is a way to offer many possible selections to the user.
Each one of the buttons must symbolize a single value.
You can use the tab key to switch from one radiobutton to another.

Syntax:
       w=Radiobutton(master,option,...)

```python
# Example:

import tkinter as tk

root = tk.Tk()

v = tk.IntVar()

tk.Label(root, 
        text="""Choose your gender:""",
        justify = tk.LEFT,
        padx = 20).pack()
tk.Radiobutton(root, 
              text="Male",
              padx = 20, 
              variable=v, 
              value=1).pack(anchor=tk.W)
tk.Radiobutton(root, 
              text="Female",
              padx = 20, 
              variable=v, 
              value=2).pack(anchor=tk.W)

root.mainloop()
```

**PRATICE EXAMPLES**

1. Display your name:

```python
from tkinter import *
def display():
    name=e1.get()
    m="Hello "+ name + ", Welcome"
    var.set(m)
    return 
window=Tk()
var=StringVar()
var.set("")
l1=Label(window,text="Enter your name:")
e1=Entry(window)
b1=Button(window,text="Display",command=display)
msg=Label(window,textvariable=var)

l1.grid(row=0,column=0)
e1.grid(row=0,column=1)
b1.grid(row=0,column=2)
msg.grid(row=1,column=0)
```

2. Concatenation of two strings:
    

```python
from tkinter import *
def add():
    s1=e1.get()
    s2=e2.get()
    try:
        x=(s1+" ")
        y=(s2)
        z=x+y
        v1.set(z)
        
    except Exception as e:
            v2.set("Error: Invalid inputs")
    return
window=Tk()
v1=StringVar()
v2=StringVar()
l1=Label(window,text="Enter your fisrt name:")
e1=Entry(window)
l2=Label(window,text="Enter your last name:")
e2=Entry(window)
l3=Label(window,text="Name after concatenation:")
e3=Entry(window,textvariable=v1)
b1=Button(window,text="CONCATENATION",command=add)
msg=Label(window,fg="red",textvariable=v2)

l1.grid(row=0,column=0)
e1.grid(row=1,column=0)
l2.grid(row=0,column=1)
e2.grid(row=1,column=1)
l3.grid(row=0,column=2)
e3.grid(row=1,column=2)
b1.grid(row=3,column=1)
msg.grid(row=2,column=0)
```

3. Addition of two numbers:

```python
from tkinter import *
def add():
    s1=e1.get()
    s2=e2.get()
    try:
        x=int(s1)
        y=int(s2)
        z=x+y
        v1.set(z)
        
    except Exception as e:
            v2.set("Error: Invalid inputs")
    return
window=Tk()
v1=StringVar()
v2=StringVar()
l1=Label(window,text="Enter first number")
e1=Entry(window)
l2=Label(window,text="Enter second number")
e2=Entry(window)
l3=Label(window,text="Result")
e3=Entry(window,textvariable=v1)
b1=Button(window,text="Add",command=add)
msg=Label(window,fg="red",textvariable=v2)

l1.grid(row=0,column=0)
e1.grid(row=1,column=0)
l2.grid(row=0,column=1)
e2.grid(row=1,column=1)
l3.grid(row=0,column=2)
e3.grid(row=1,column=2)
b1.grid(row=3,column=1)
msg.grid(row=2,column=0)
```
